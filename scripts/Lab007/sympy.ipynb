```python
x = sp.Function('x')
w0 = sp.Symbol('omega_0', real = True, positive = True)
t = sp.Symbol('t')
x0 = sp.Symbol('x_0', real = True)
v0 = sp.Symbol('v_0', real = True)

x(t), w0, t, x0, v0
```

```python
x(t).diff(t, t)
```

```python
-w0**2 * x(t)
```

```python
eq = sp.Eq(x(t).diff(t, t), -w0**2 * x(t))
eq
```

```python
x(t).diff(t).subs(t, 0)
```

```python
sol = sp.dsolve(eq, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
```

```python
sol.simplify()
```

```python
sol.lhs, sol.rhs
```

```python
sol.rhs.subs({x0: 1, v0: -1, w0: 2})
```

```python
sp.plot(sol.rhs.subs({x0: 1, v0: -1, w0: 2}), (t, -5, 5))
```

```python
solf = sp.lambdify(t, sol.rhs.subs({x0: 1, v0: -1, w0: 2}))
```

```python
solf(2), type(solf(2))
```

```python
ts = np.linspace(-5, 5, 1000)
plt.plot(ts, solf(ts))
plt.grid()
```
