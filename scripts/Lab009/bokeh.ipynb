```python
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import row, column, gridplot, layout
from bokeh.models import Slider, Div
from bokeh.util.hex import hexbin
from bokeh.transform import linear_cmap
from bokeh.palettes import all_palettes

output_notebook()
```

```python
ts1 = np.linspace(1, 10, 1000)
ys1 = np.sin(ts1)
ys2 = np.cos(ts1)
ys3 = np.sin(ts1) + np.cos(ts1)
```

```python
fig = figure(sizing_mode = 'stretch_width',
             aspect_ratio = 3,
             title = 'Mój mega wykres!',
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$')
fig.toolbar.logo = None
fig.toolbar.autohide = True
fig.grid.grid_line_dash = (5, 5)

fig.line(ts1, ys1, color = 'pink', line_width = 3, legend_label = 'sin')
fig.line(ts1, ys2, color = 'magenta', line_width = 3, legend_label = 'cos')
fig.line(ts1, ys3, color = 'orange', line_width = 3, legend_label = 'sin + cos')
fig.circle(ts1[::20], ys3[::20], color = 'orange', legend_label = 'sin + cos', radius = 0.06, alpha = 0.5)

show(fig)
```

```python
fig1 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$',
             sizing_mode = 'stretch_width')
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.grid.grid_line_dash = (5, 5)
fig1.line(ts1, ys1, color = 'pink', line_width = 3, legend_label = 'sin')

fig2 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$',
             sizing_mode = 'stretch_width')
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.grid.grid_line_dash = (5, 5)
fig2.line(ts1, ys2, color = 'magenta', line_width = 3, legend_label = 'cos')

fig3 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$',
             sizing_mode = 'stretch_width')
fig3.toolbar.logo = None
fig3.toolbar.autohide = True
fig3.grid.grid_line_dash = (5, 5)
fig3.line(ts1, ys3, color = 'orange', line_width = 3, legend_label = 'sin + cos')

fig4 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$',
             sizing_mode = 'stretch_width')
fig4.toolbar.logo = None
fig4.toolbar.autohide = True
fig4.grid.grid_line_dash = (5, 5)
fig4.circle(ts1[::20], ys3[::20], color = 'orange', legend_label = 'sin + cos', radius = 0.06, alpha = 0.5)

show(row(fig1, fig2, fig3, fig4, sizing_mode = 'stretch_width'))
```

```python
fig1 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$')
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.grid.grid_line_dash = (5, 5)
fig1.line(ts1, ys1, color = 'pink', line_width = 3, legend_label = 'sin')

fig2 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$')
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.grid.grid_line_dash = (5, 5)
fig2.line(ts1, ys2, color = 'magenta', line_width = 3, legend_label = 'cos')

fig3 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$')
fig3.toolbar.logo = None
fig3.toolbar.autohide = True
fig3.grid.grid_line_dash = (5, 5)
fig3.line(ts1, ys3, color = 'orange', line_width = 3, legend_label = 'sin + cos')

fig4 = figure(aspect_ratio = 1,
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$')
fig4.toolbar.logo = None
fig4.toolbar.autohide = True
fig4.grid.grid_line_dash = (5, 5)
fig4.circle(ts1[::20], ys3[::20], color = 'orange', legend_label = 'sin + cos', radius = 0.06, alpha = 0.5)

show(gridplot([[fig1, fig2], [fig3, fig4]], width = 300))
```

```python
fig = figure(title = 'Mój mega wykres!',
             x_axis_label = '$$t$$',
             y_axis_label = '$$\Omega$$',
             width = 400,
             aspect_ratio = 1)
fig.toolbar.logo = None
fig.toolbar.autohide = True
fig.grid.grid_line_dash = (5, 5)

fig.line(ts1, ys3, color = 'orange', line_width = 3)
c = fig.circle(ts1[::20], ys3[::20], color = 'orange', radius = 0.1, alpha = 0.5)

s1 = Slider(start = 0, end = 1, step = 0.01, value = 0.5, title = 'alpa', sizing_mode = 'stretch_width')
s1.js_link('value', c.glyph, 'fill_alpha')
s2 = Slider(start = 0.5, end = 2, step = 0.1, value = 0.1, title = 'radius', sizing_mode = 'stretch_width')
s2.js_link('value', c.glyph, 'radius')

show(row(column(s1, s2, width = 200), fig))
```

```python
n = 100000

xs = np.random.normal(0, 1, n)
ys = np.random.normal(0, 1, n)

hb = hexbin(xs, ys, 0.2)

fig = figure(width = 400, aspect_ratio = 1)
fig.background_fill_color = all_palettes['Turbo'][256][0]
fig.grid.visible = False

cmap = linear_cmap('counts', 'Turbo256', 0, np.max(hb['counts']))
fig.hex_tile(source = hb, size = 0.2, color = cmap, line_color = None)

s1 = Slider(start = 100, end = 1000000, step = 100, value = 100000, title = 'N', sizing_mode = 'stretch_width')

show(column(Div(text = 'To jest rozklad normalny!'), row(column(s1, width = 200), fig)))
```
