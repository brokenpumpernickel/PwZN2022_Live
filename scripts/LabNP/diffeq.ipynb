```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

$$m\frac{d^2x}{dt^2} = QE - B v$$
$$\frac{d^2x}{dt^2} = qE - b v$$

$$\frac{dv}{dt} = qE - b v$$
$$\frac{dx}{dt} = v$$

$$\vec{y} = [v, x]$$
$$\frac{d\vec{y}}{dt} = [qE - b v, v]$$

```python
q = 1
E = 1
b = 10

def func(y, t):
    v, x = y
    return q * E - b * v, v

v0 = 1
x0 = 0

ts1 = np.linspace(0, 1, 1000)
res1 = odeint(func, (v0, x0), ts1)
```
